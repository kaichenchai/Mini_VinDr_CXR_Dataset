```python
import matplotlib.pyplot as plt
import cv2
```

```python
bit_2 = cv2.imread("2-bit.png", cv2.IMREAD_GRAYSCALE)
#bit_2 = bit_2 / (2**2) * (2**16)

bit_4 = cv2.imread("4-bit.png", cv2.IMREAD_GRAYSCALE)
#bit_4 = bit_4 / (2**4) * (2**16)

bit_8 = cv2.imread("8-bit.png", cv2.IMREAD_GRAYSCALE)
#bit_8 = bit_8 / (2**8) * (2**16)

bit_16 = cv2.imread("16-bit.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(bit_2, cmap="grey")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 8))

for ax in axes.ravel():
    ax.set_axis_off()
    
axes[0].set_title("2-Bit Image")
axes[0].imshow(bit_2, cmap='gray')

axes[1].set_title("4-Bit Image")
axes[1].imshow(bit_4, cmap='gray')

axes[2].set_title("8-Bit Image")
axes[2].imshow(bit_8, cmap='gray')

axes[3].set_title("16-Bit Image")
axes[3].imshow(bit_16, cmap='gray')
```
