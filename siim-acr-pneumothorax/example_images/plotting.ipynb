```python
import matplotlib.pyplot as plt
import cv2
```

```python
original = cv2.imread("original.png", cv2.IMREAD_GRAYSCALE)
clahe = cv2.imread("clahe.png",cv2.IMREAD_GRAYSCALE)
clahe_bs = cv2.imread("clahe_bs.png", cv2.IMREAD_GRAYSCALE)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 8))

for ax in axes.ravel():
    ax.set_axis_off()
    
axes[0].set_title("Original")
axes[0].imshow(original, cmap='gray')

axes[1].set_title("CLAHE")
axes[1].imshow(clahe, cmap='gray')

axes[2].set_title("CLAHE and Bone Suppression")
axes[2].imshow(clahe_bs, cmap='gray')
```
