```python
from YOLOv8_Explainer import yolov8_heatmap, display_images
```

```python
model = yolov8_heatmap(
    weight="best.pt", 
        conf_threshold=0.4, 
        method = "EigenCAM",
        layer = [10, 12, 14, 16, 18, -3],
        ratio=0.02,
        show_box=True,
        renormalize=False,
)
```

```python
imagelist = model(
    img_path="images/",
    )

len(imagelist)
```

```python
display_images(imagelist)
```
