```python
from ultralytics import YOLO
from IPython.display import display, Image
import wandb
from wandb.integration.ultralytics import add_wandb_callback
import os
```

```python
print(os.getcwd())
try:
  os.chdir("Mini_VinDr_CXR_Dataset")
except FileNotFoundError as err:
  print(f"[LIKELY ALREADY IN FOLDER]{err}")
else:
  print(os.getcwd())
  print(os.listdir("/content/Mini_VinDr_CXR_Dataset"))
```

```python
wandb.init(project="YOLOv8_VinDR", job_type="training", name = "YOLOv8s_miniVinDR",
config={
    "epochs": 10,
    "dataset": "mini_VinDR",
    "model": "YOLOv8s",
    "image_size": 1024
}
)

# Load a pretrained YOLO model (recommended for training)
model = YOLO("yolov8s.pt")
dataset_name = "model.yaml"

# Add the Weights & Biases callback to the model.
# This will work for training, evaluation and prediction
add_wandb_callback(model, enable_model_checkpointing=True)

# Train the model using the VinDR dataset for 5 epochs
# Results after evaluating the validation batch would be logged
# to a W&B table at the end of each epoch
model.train(project = "Mini_VinDr_YOLOv8", data="/content/Mini_VinDr_CXR_Dataset/model.yaml", epochs=10, imgsz=1024)
```
